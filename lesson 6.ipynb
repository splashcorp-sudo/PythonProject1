```python
while True:
    user_input= input('say something')
    if user_input=='exit':
        break
    elif user_input=='skip':
        print('skipping')
        continue
    elif len(user_input)>10:
        print('your input is too long')
    else:
        print('ok')
print('goodbye')
```

```python
a = 1
b = 5
c = 4
d = 7
y = 1

main_number = 47

def calc(numb):
    if y == 0:
        return numb
    else:
        return numb + main_number
calc(a)
calc(b)
calc(c)
calc(d)
```

```python
def print_words (first, second, third, fourth, fifth):
    print(f'the first is {first}, second is {second}, third is {third}, fourth is {fourth}, {fifth}')
print_words('one', 'two', 'three', 'four', 'five')
print_words(second='two', fourth='four', third='three', first='one', fifth='five')
```

возведение в степень

```python
def power (nember, degree=2 ):
    return nember ** degree
print(power(3))
print(power(2,3))
print(power(4,3))
```

```python
def sum_all(*args): # * означает что принимаем неограниченное число не именнованных аргументов
                    # ** означает что принимаем неограниченное число именнованных аргументов
    return sum(args)
print(sum_all(1,2,3,4,5))
```

```python
def price_list(**kwargs): # ** означает что принимаем неограниченное число именнованных аргументов
    print (kwargs)
price_list(iphone=2500, laprop=1500, samsung=900)
```

```python
def price_list(**kwargs):
    for title,price in kwargs.items():
        print(f'product {title} price is {price}')
price_list(iphone=2500, laprop=1500, samsung=900)
```

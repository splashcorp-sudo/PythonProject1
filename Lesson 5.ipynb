```python
citys=['Moscow', 'New York', 'San Francisco', 'Kiev', 'San Diego' ]
for city in citys:
    print(city)
```

```python
citys=['Moscow', 'New York', 'San Francisco', 'Kiev', 'San Diego' ]
for city in citys:
    if city.startswith('S'):
        print(city)
```

```python
scale= {'Tim': 154, 'Alex': 205, 'Nikita': 130}
for scal in scale:
    print(scal)
```

```python
scale= {'Tim': 154, 'Alex': 205, 'Nikita': 130}
print(scale.values())
for scal in scale.values():
    print(scal)
```

```python
scale= {'Tim': 154, 'Alex': 205, 'Nikita': 130}
for scal in scale:
    print(f'{scal}:{scale[scal]}')
```

```python
scale= {'Tim': 154, 'Alex': 205, 'Nikita': 130}
for name, height in scale.items():
    print(f'{name}:{height}')
```
